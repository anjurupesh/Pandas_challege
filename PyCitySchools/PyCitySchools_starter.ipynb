```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
total_schools= school_data_df["school_name"].count()

total_schools
```

```python
total_students= school_data_complete_df ['student_name'].count()
total_students
```

```python
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score = round(average_math_score,2)
average_math_score
```

```python
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score = round(average_reading_score,2)
average_reading_score
```

```python
overall_passing_score = (average_math_score+average_reading_score)/2
overall_passing_score
```

```python
number_passing_student = school_data_complete_df[school_data_complete_df["math_score"] > 70] 
number_passing_student = number_passing_student["school_name"].count() 
perc_passing_student = (number_passing_student/total_students)*100
perc_passing_student= round(perc_passing_student,2)
perc_passing_student
```

```python
number_passing_student = school_data_complete_df[school_data_complete_df["reading_score"] > 70] 
number_passing_student = number_passing_student["school_name"].count() 
perc_passing_student = (number_passing_student/total_students)*100
perc_passing_student= round(perc_passing_student,2)
perc_passing_student
```

```python
summary_table = [{"Number of Schools":total_schools, "Number of students":total_students, "Total Budget":total_budget,"Average Maths Score":average_math_score,"Average Reading Score":average_reading_score,"Total Average Passing Score": overall_passing_score,"Pass Students Percentage": perc_passing_student}]
summary_table_df = pd.DataFrame(summary_table)
summary_table_df
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_table_df = school_data_complete_df.groupby("school_name")

school_type_df = [s_type[0] for s_type in school_table_df["type"].unique()]

student_number_schoolwise_df = school_table_df["Student ID"].count()

budget_schoolwise_df = school_table_df["budget"].mean() 

budget_per_student_schoolwise_df = (school_table_df["budget"].mean())/(school_table_df["Student ID"].count())

pass_math_school = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]
math_group = pass_math_school.groupby("school_name")

pass_reading_school = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]
reading_group = pass_reading_school.groupby("school_name")

average_mathsscore_schoolwise_df = school_table_df["math_score"].mean()
average_readingscore_schoolwise_df = school_table_df["reading_score"].mean()

perc_passing_mathsscore_schoolwise_df = (math_group["student_name"].count())/(school_table_df["student_name"].count())*100

perc_passing_reading_schoolwise_df = (reading_group["student_name"].count())/(school_table_df["student_name"].count())*100
average_overall_passing_df = (perc_passing_mathsscore_schoolwise_df+perc_passing_reading_schoolwise_df)/2


school_summary_table= pd.DataFrame({"School Type":school_type_df, "Student Number": student_number_schoolwise_df,
                                   "Budget":budget_schoolwise_df,"Budget per Student":budget_per_student_schoolwise_df,
                                    "Average Maths Score":average_mathsscore_schoolwise_df,"Average Reading Score":average_readingscore_schoolwise_df,
                                    "Percent Passing Maths":perc_passing_mathsscore_schoolwise_df,"Percent Passing Reading":perc_passing_reading_schoolwise_df,
                                    "Percent Overall Passing":average_overall_passing_df})

school_summary_table["Budget"] = school_summary_table["Budget"].astype(float)
school_summary_table["Budget per Student"] = school_summary_table["Budget per Student"].astype(float)

school_summary_table.head(15)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python

Ninth_grade = school_data_complete_df.loc[school_data_complete_df["grade"] == "9th"]
Ninth_grade_group = Ninth_grade.groupby("school_name")
Ninth_grade_group_avg = Ninth_grade_group.mean()

Tenth_grade = school_data_complete_df.loc[school_data_complete_df["grade"] == "9th"]
Tenth_grade_group = Tenth_grade.groupby("school_name")
Tenth_grade_group_avg = Tenth_grade_group.mean()

Eleventh_grade = school_data_complete_df.loc[school_data_complete_df["grade"] == "9th"]
Eleventh_grade_group = Eleventh_grade.groupby("school_name")
Eleventh_grade_group_avg = Eleventh_grade_group.mean()

Twelth_grade = school_data_complete_df.loc[school_data_complete_df["grade"] == "9th"]
Twelth_grade_group = Twelth_grade.groupby("school_name")
Twelth_grade_group_avg = Twelth_grade_group.mean()


grade_summary_ni_te = pd.merge(Ninth_grade_group_avg, Tenth_grade_group_avg, on="school_name", suffixes=("_ni", "_te"))
grade_summary_el_tw = pd.merge(Eleventh_grade_group_avg, Twelth_grade_group_avg, on="school_name", suffixes=("_el", "_tw"))
grade_summary_total = pd.merge(grade_summary_ni_te, grade_summary_el_tw, on="school_name")

math_score = grade_summary_total[["math_score_ni", "math_score_te", "math_score_el", "math_score_tw"]]
renamed_math_score = math_score.rename(columns={"math_score_ni": "9th", "math_score_te": "10th", "math_score_el": "11th", "math_score_tw": "12th"})
renamed_math_score.index.name = None
renamed_math_score
```
